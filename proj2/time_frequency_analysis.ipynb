```python
"""
Best,
Sam
"""

import mne
import os

edf_path = r"C:\Users\samue\CC_dataset\music\drive\sub-02\eeg\sub-02_task-run3_eeg.edf"

raw = mne.io.read_raw_edf(edf_path, preload=True)
print(raw.info)
raw.plot(duration=10, n_channels=12)

```

```python
# PSD using Welch's
import matplotlib.pyplot as plt

data = raw.get_data()
psds, freqs = mne.time_frequency.psd_array_welch(data, sfreq=1000, fmin=0, fmax=500, n_fft=256)

# Plot
plt.figure(figsize=(10, 6))
plt.semilogy(freqs, psds.T)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density (dB/Hz)')
plt.title('Power Spectral Density (PSD)')
plt.show()
```
